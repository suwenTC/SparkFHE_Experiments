```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

```python
sns.set(rc={'figure.figsize': (16, 16)},style='darkgrid', font_scale=1.5)
```

# Docker

```python
columns = ['time(s)', 'cpu(%)', 'mem(%)']
```

```python
docker_basic_op_4 = pd.read_csv("docker/basic_op_4.csv", header=None)
docker_basic_op_4.columns = columns

k8s_basic_op_4 = pd.read_csv("k8s/basic_op_4.csv", index_col=0)
k8s_basic_op_4.columns = columns

docker_mix_4 = pd.read_csv("docker/mix_4.csv", header=None)
docker_mix_4.columns = columns

k8s_mix_4 = pd.read_csv("k8s/mix_4.csv", index_col=0)
k8s_mix_4.columns = columns
```

```python
_=sns.lineplot(x='time(s)', y='cpu(%)', data=docker_basic_op_4)
_=sns.lineplot(x='time(s)', y='cpu(%)', data=k8s_basic_op_4)
_=sns.lineplot(x='time(s)', y='cpu(%)', data=docker_mix_4)
_=sns.lineplot(x='time(s)', y='cpu(%)', data=k8s_mix_4)

_=plt.legend(['docker', 'k8s', 'docker mix', 'k8s mix'])
_=plt.title('4 Jobs')
```

```python
_=sns.lineplot(x='time(s)', y='mem(%)', data=docker_basic_op_4)
_=sns.lineplot(x='time(s)', y='mem(%)', data=k8s_basic_op_4)
_=sns.lineplot(x='time(s)', y='mem(%)', data=docker_mix_4)
_=sns.lineplot(x='time(s)', y='mem(%)', data=k8s_mix_4)

_=plt.legend(['docker', 'k8s', 'docker mix', 'k8s mix'])
_=plt.title('4 Jobs')
```
